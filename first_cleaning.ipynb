```python
df = df.drop(['user_location', 'user_reactions', 'scraped_at'], axis=1)
df.head()
```

```python
# View unique usernames
unique_usernames = df['username'].unique()
print(unique_usernames)

# Get count of unique usernames
unique_count = df['username'].nunique()
print(f"Count of unique usernames: {unique_count}")
```

```python
username_counts = df['username'].value_counts()
print(username_counts)
```

```python
all_duplicates = df[df['post_id'].duplicated(keep=False)]
# print(all_duplicates['all_duplicates'].unique())
all_duplicates
```

```python
general_discussions = df[df['forum_category'] == 'General discussions about fibromyalgia']
count = len(general_discussions)
count
```

```python
# View unique posts
unique_posts = general_discussions['post_id'].unique()
print(unique_posts)

# Get count of unique posts
unique_count = general_discussions['post_id'].nunique()
print(f"Count of unique posts: {unique_count}")
```

```python
username_counts = general_discussions['post_id'].value_counts()
print(username_counts)
```

```python
duplicates = general_discussions[general_discussions['post_id'].duplicated(keep=False)]
# print(duplicates['post_id'].unique())
duplicates
```

```python
duplicates_all = df[(df['post_id'].duplicated(keep=False)) &
                                 (df['thread_views'] == '–') & (df['thread_replies'] == '–')]
duplicates_all
```

```python
df = df[~((df['post_id'].duplicated(keep=False)) &
          (df['thread_views'] == '–') & (df['thread_replies'] == '–'))]
```
